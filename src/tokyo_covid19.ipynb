```python
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import numpy

df = pd.read_csv("https://stopcovid19.metro.tokyo.lg.jp/data/130001_tokyo_covid19_patients.csv", header=0)
countByDate = df.groupby('公表_年月日').count()
datacumsum = numpy.cumsum(countByDate,axis=0)
sns.set()
plt.plot(datacumsum.iloc[:,0])
             
```
